### Query 6.3 
Cluster the 30 stocks according to their mean annual Volatilities and Returns using K-means clustering. Identify the optimum number of clusters using the Elbow curve method

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
data = returns[['Returns', 'Volatility']].values
data
```

```python
# Transforming data
mms = MinMaxScaler()
mms.fit(data)
transformedData = mms.transform(data)
```

```python
# calculating Inertia avrage dist. b/w the cluster center and its points
Sum_of_squared_distances = []
K = range(1, 31)
for k in K:
    km = KMeans(n_clusters = k)
    km = km.fit(transformedData)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
# Plotting the Elbow Method Graph
fig, ax = plt.subplots(figsize=(20, 10))

ax.plot(K, Sum_of_squared_distances, 'bx-')

plt.xticks(K)
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')

plt.title('Elbow Method For Optimal k')
plt.show()
```

### Query 6.4
Prepare a separate Data frame to show which stocks belong to the same cluster

```python
# Using k=6
k = 6
km = KMeans(n_clusters = k)
km = km.fit(transformedData)
```

```python
returns['Labels'] = km.labels_
centers = km.cluster_centers_
y_kmeans = km.predict(transformedData)
```

```python
plt.style.use('bmh')

fig, ax = plt.subplots(figsize=(20, 15))
ax.scatter(transformedData[:, 0], 
            transformedData[:, 1], 
            c = y_kmeans, 
            s = 50, 
            cmap = 'viridis'
            )

ax.scatter(centers[:, 0], 
           centers[:, 1], 
           c = 'black', 
           s = 10000, 
           alpha = 0.05
           )

font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 10,
        }

for x, y, name in zip(transformedData[:, 0], transformedData[:, 1], returns.index):
    ax.text(x, 
            y, 
            name, 
            fontdict=font,
            # verticalalignment='center', 
            # horizontalalignment='center',
            rotation=0
            )


font = {'family': 'serif',
        'color':  'gray',
        'weight': 'normal',
        'size': 25,
        }

for i, cord in enumerate(centers):
    ax.text(cord[0], 
            cord[1], 
            f"C{i}",
            horizontalalignment = 'center',
            verticalalignment = 'center', 
            fontdict = font,             
            rotation = 0, 
            alpha = 0.3
            )
plt.xlabel('Scaled Returns')
plt.ylabel('Scaled Volatility')
plt.title('K-Means Clustering')
plt.show()
```
