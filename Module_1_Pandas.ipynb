### Query 1.5
Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days. Calculate the average price AND the profit/loss percentages over the course of last - 1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year. {Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }

```python
def avg_last_N_days(N):
  '''
  N = Number of days 
  '''

  return np.sum(pd.Series(df['Average Price'].tail(N))) / N
```

```python
def prof_loss(N):
  '''
  N = Number of days 
  '''
  diff = df['Close Price'].tail(N).iloc[N-1] - df['Close Price'].tail(N).iloc[0]
  if diff < 0:
    loss_per = -(diff / (df['Close Price'].tail(N).iloc[N-1])) * 100
    return f'{loss_per}% loss'
  else:
    prof_per = (diff / (df['Close Price'].tail(N).iloc[N-1])) * 100
    prof_per
    return f'{prof_per}% profit'
```

```python
# required outputs
print('Average prices for last N days are as follows: ')
print('For last 1 Week: ', avg_last_N_days(5))
print('For last 2 Weeks: ', avg_last_N_days(10))
print('For last 1 Month: ', avg_last_N_days(20))
print('For last 3 Months: ', avg_last_N_days(60))
print('For last 6 Months: ', avg_last_N_days(120))
print('For last 1 Year: ', avg_last_N_days(240))
print('*'*100)
print('Profit/Loss Percentages for last N days are as follows: ')
print('For last 1 Week: ', prof_loss(5))
print('For last 2 Weeks: ', prof_loss(10))
print('For last 1 Month: ', prof_loss(20))
print('For last 3 Months: ', prof_loss(60))
print('For last 6 Months: ', prof_loss(120))
print('For last 1 Year: ', prof_loss(240))
```

### Query 1.7
Add another column 'Trend' whose values are: 

*   'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5  
*   'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1
*   'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1
*   'Positive' for 'Day_Perc_Change' in between 1 and 3
*   'Negative' for 'Day_Perc_Change' in between -1 and -3
*   'Among top gainers' for 'Day_Perc_Change' in between 3 and 7
*   'Among top losers' for 'Day_Perc_Change' in between -3 and -7
*   'Bull run' for 'Day_Perc_Change' >7
*   'Bear drop' for 'Day_Perc_Change' <-7
 

```python
if ((df['Day_Perc_Change'] >= -0.5) & (df['Day_Perc_Change'] <= 0.5)).all():
    df['Trend'] = 'Slight or No change'
if ((df['Day_Perc_Change'] >= 0.5) & (df['Day_Perc_Change'] <= 1)).all():
    df['Trend'] = 'Slight positive'
if ((df['Day_Perc_Change'] <= -0.5) & (df['Day_Perc_Change'] >= -1)).all():
    df['Trend'] = 'Slight negative'
if ((df['Day_Perc_Change'] >= 1) & (df['Day_Perc_Change'] <= 3)).all():
    df['Trend'] = 'Positive' 
if ((df['Day_Perc_Change'] <= -1) & (df['Day_Perc_Change'] >= -3)).all():
    df['Trend'] = 'Negative'
if ((df['Day_Perc_Change'] >= 3) & (df['Day_Perc_Change'] <= 7)).all():
    df['Trend'] = 'Among top gainers'
if ((df['Day_Perc_Change'] <= -3) & (df['Day_Perc_Change'] >= -7)).all():
    df['Trend'] = 'Among top losers'
if (df['Day_Perc_Change'] > 7).all():
    df['Trend'] = 'Bull run' 
if (df['Day_Perc_Change'] < -7).all():
    df['Trend'] = 'Bear drop' 
```
